```julia
using DelimitedFiles
using HDF5
using Plots
```

```julia
n = 50
dats = fill(NaN, (4,3,n,10))
for i in 1:10
    f = h5read("../data/base_single/perplex_out_$i.h5", "results")
    dats[:,:,:,i] .= f[:,:,1:n]
end
```

```julia
size(dats)
```

```julia
plot(vec(dats[2,2,:,:]),vec(dats[3,2,:,:]),vec(dats[4,2,:,:]),
    marker_z=repeat(1:10,inner=n),
    seriestype=:scatter,zlims=(1.4,1.75),
				xlims=(2250,3000), ylims=(4.75,8.25), markersize=3, markerstrokewidth=0)
```
